```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras import Sequential
from tensorflow.keras .layers import Dense, Dropout, Activation, Embedding, SimpleRNN, LSTM
from tensorflow.keras.preprocessing.sequence import pad_sequences   
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
from tensorflow.keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words = 500)
```

```python
x_train = x_train[:2500]
x_test = x_test[:2500]
y_train = y_train[:2500]
y_test = y_test[:2500]
```

```python
y_train
```

```python
lengths = [len(line) for line in x_train]
lengths
```

```python
plt.hist(lengths, bins = 100)
```

```python
x_train_seq = pad_sequences(x_train, maxlen = 200)
x_test_seq = pad_sequences(x_test, maxlen = 200)
```

```python
print(x_train_seq.shape)
print(x_test_seq.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()
model.add(SimpleRNN(8, input_shape = (200,1)))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', metrics=['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train_seq, y_train, epochs = 10, batch_size = 64, validation_data=(x_test_seq, y_test))
```

```python
x_train_oh = to_categorical(x_train_seq)
x_test_oh = to_categorical(x_test_seq)
```

```python
x_train_oh
```

```python
x_train_oh.shape
```

```python
model = Sequential()
model.add(SimpleRNN(8, input_shape = (200, 500)))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train_oh, y_train, epochs = 10, batch_size=64, validation_data=(x_test_oh, y_test))
```

```python
model = Sequential()
model.add(LSTM(8, input_shape=(200,500)))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train_oh, y_train, epochs = 10, batch_size = 64, validation_data=(x_test_oh, y_test))
```

```python
model = Sequential()
model.add(Embedding(500, 16, input_length=200))
model.add(LSTM(8))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
hist = model.fit(x_train_seq, y_train, epochs = 10, batch_size = 64, validation_data=(x_test_seq, y_test))
```

### reuter_lstm

```python
from tensorflow.keras.datasets import reuters

(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words = 1000, test_split=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x_train
```

```python
np.unique(y_train)
```

```python
lengths = [len(line) for line in x_train]
lengths
```

```python
plt.hist(lengths, bins = 100)
```

```python
x_train_seq = pad_sequences(x_train, maxlen=100)
x_test_seq = pad_sequences(x_test, maxlen=100)
```

```python
print(x_train_seq.shape)
print(x_test_seq.shape)
```

```python
y_train_oh = to_categorical(y_train)
y_test_oh = to_categorical(y_test)
```

```python
y_train_oh.shape
```

```python
# x_train_oh = to_categorical(x_train_seq)
# x_test_oh = to_categorical(x_test_seq)
```

```python
# x_test_oh.shape
```

```python
model = Sequential()
model.add(Embedding(1000, 16, input_length=100))
model.add(LSTM(8))
model.add(Dense(46, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train_seq, y_train_oh, epochs = 30, batch_size = 20, validation_data=(x_test_seq, y_test_oh))
```

```python
scaler = StandardScaler()
scaler.fit(x_train_seq)
x_train_sc = scaler.transform(x_train_seq)
x_test_sc = scaler.transform(x_test_seq)
```

```python
model = Sequential()
model.add(Embedding(1000, 16, input_length=100))
model.add(LSTM(8))
model.add(Dense(46, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train_sc, y_train_oh, epochs = 30, batch_size = 20, validation_data=(x_test_sc, y_test_oh))
```

### Timeless_LSTM

```python
ret = requests.get('https://poloniex.com/public?command=returnChartData&currencyPair=USDT_BTC&start=1405699200&end=9999999999&period=86400')
```

```python
df = pd.DataFrame(ret.json())
df
```

```python
scaler = MinMaxScaler()
```

```python
price = scaler.fit_transform(df)[:,4]
price = list(price)
```

```python
len(price)
```

```python
window_size = 10
X = []
Y = []

for i in range(len(price) - window_size):
    X.append([price[i+j] for j in range(window_size)])
    Y.append(price[window_size + i])
```
