```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.utils import *
from sklearn.preprocessing import *
```

```python
df = pd.read_csv('../Data/eighthr.data', header = None)
df
```

```python
# neural network를 구성해서 예측하는 모델, LSTM으로 예측하는 모델을 각각 생성하고 평가하시오
# target : 73번 칼럼
# object => numeric
# scaling
```

```python
df.info()
```

```python
df = df.iloc[:, 1:].apply(pd.to_numeric, errors = 'coerce')
df.head()
```

```python
df1 = df.dropna(axis = 0)
```

```python
df1
```

```python
scaler = MinMaxScaler()
df1_sc = scaler.fit_transform(df1)
df1_sc
```

```python
y = df1_sc[:, -1]
x = df1_sc[:, :-1]
```

```python
x = x.reshape(1847, 72, 1)
```

```python
y.shape
```

```python
model = Sequential()
model.add(LSTM(128, input_shape = (72, 1)))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x, y, epochs = 20, batch_size = 10, validation_split = 0.2)
```
