```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.utils import to_categorical
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# convolution layer를 이용해서 multiclassification nural network model을 생성하고 테스트 평가하시오
```

```python
# input data shape => (데이터수, row num, col num, depth)
# multiclassification => target : onehot 
```

```python
# 스케일링

x_train = x_train / 255
x_test = x_test / 255
```

```python
x_train
```

```python
y_train_oh = tf.keras.utils.to_categorical(y_train)
y_test_oh = tf.keras.utils.to_categorical(y_test)
```

```python
y_train_oh.shape
```

```python

model = Sequential()
model.add(Conv2D(16, kernel_size=4, activation = 'relu', input_shape = (32, 32, 3), padding = 'same'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Conv2D(32, kernel_size = 4, activation = 'relu', padding = 'same'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Conv2D(64, kernel_size = 4, activation = 'relu', padding = 'same'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Flatten())
model.add(Dense(512, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
model.summary()
```

```python

model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
hist = model.fit(x_train, y_train_oh, validation_split=0.2, epochs = 10, batch_size = 10)
```

```python
model.evaluate(x_test, y_test_oh)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```
