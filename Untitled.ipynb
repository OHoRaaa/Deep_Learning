```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
```

```python
df = pd.read_csv('./Data/housing.csv', header = None, delim_whitespace=True)
df
```

```python
y = df.loc[:, 13]
x = df.loc[:, :12]
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=111)
```

```python
model = Sequential()
model.add(Dense(50, input_dim = 13, activation = 'relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer = 'adam')
```

```python
hist = model.fit(x_train, y_train, validation_split=0.2, epochs = 100, batch_size = 20)
```

```python
model.evaluate(x_test, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_sc = scaler.transform(x_train)
x_test_sc = scaler.transform(x_test)
```

```python
model = Sequential()
model.add(Dense(50, input_dim = 13, activation = 'relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer = 'adam')
```

```python
hist = model.fit(x_train_sc, y_train, validation_split=0.2, epochs = 100, batch_size = 20)
```

```python
model.evaluate(x_test_sc, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
preds = model.predict(x_test_sc)
preds
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, preds)
```

```python
df = pd.read_csv('./Data/adult.txt', header=None, names = ['age', 'workclass','fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race'\
, 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', '50k'])

df
```

```python
df.info()
```

```python
y = df['50k'].apply(lambda x : 1 if x == ' >50K' else 0)
y
```

```python
x = df.loc[:, :'native-country']
x = pd.get_dummies(x)
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=111)
```

```python
model = Sequential()
model.add(Dense(200, input_dim = 108, activation = 'relu'))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(60, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x_train, y_train, validation_split=0.2, epochs = 100, batch_size = 30)
```

```python
model.evaluate(x_test, y_test)
```

```python
hist = model.fit(x_train, y_train, validation_split=0.2, epochs = 100, batch_size = 30)
```

```python
plt.plot(hist.history['loss'], c = 'r')
plt.plot(hist.history['val_loss'], c = 'b')
```
